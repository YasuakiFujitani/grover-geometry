### Section 3

We calculate edge Ricci curvature on undirected graph, which is defined in 

[1] Taiki Yamada "AN ESTIMATE OF THE FIRST NON-ZERO EIGENVALUE OF THE LAPLACIAN BY THE RICCI CURVATURE ON EDGES OF GRAPHS", Osaka Journal of Mathematics. 57(1) (2020) P.151-P.163.

We calculate examples in [1].

Example 4.1 in [1]

```python
import networkx as nx
from edge_ricci_undirected import UndirectedEdgeGraph
G = nx.complete_graph(4)
G_D = UndirectedEdgeGraph(G.edges)
G_D.plot()
print(G_D.compute_ricci_curvature_between_edge_undirected())
```

```python
G = nx.complete_graph(5)
G_D = UndirectedEdgeGraph(G.edges)
G_D.plot()
print(G_D.compute_ricci_curvature_between_edge_undirected())
```

```python
G = nx.complete_graph(6)
G_D = UndirectedEdgeGraph(G.edges)
G_D.plot()
print(G_D.compute_ricci_curvature_between_edge_undirected())
```

```python
G = [(0,1),(1,2),(2,0)]
G_D = UndirectedEdgeGraph(G)
G_D.plot()
print(G_D.get_distribution((0,1)))
```

```python
G = [(0,1),(1,2),(2,3),(3,0)]
G_D = UndirectedEdgeGraph(G)
G_D.plot()
print(G_D.compute_ricci_curvature_between_edge_undirected())
```

```python
G = [(0,1),(1,2),(2,3),(3,4),(4,0)]
G_D = UndirectedEdgeGraph(G)
G_D.plot()
print(G_D.compute_ricci_curvature_between_edge_undirected())
```

### Section 4

We calculate the Ricci curvature on derected graphs, which is defined in

[2] Yamada, Taiki. "THE RICCI CURVATURE ON DIRECTED GRAPHS." Journal of the Korean Mathematical Society 56.1 (2019): 113-125.

We calculate examples in [2].

Example 3.3 in [2]

```python
import networkx as nx
from directed_ricci import get_distribution, distance_between_nodes, compute_ricci_curvature_between_nodes_directed
G = [(0,1),(1,2),(2,3),(3,4),(4,0)]
G_D = nx.DiGraph(G)
print(compute_ricci_curvature_between_nodes_directed(G_D))
```
