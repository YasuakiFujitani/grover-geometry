### Example 1

```python
from hyper_graph import HyperGraph
H = HyperGraph([(0,1),(1,2),(2,0),(0,1,2)])
H.plot()
```

### Example 2

```python
from hyper_graph import line
G = line(3)
H_line = HyperGraph(G.edges)
H_line.plot()
```

We calculate $c_H$.

```python
print("Distance between [a,b] and [b,c] is", H_line.distance((1,(0,1)),(2,(1,2)))) #Figure 1
print("Distance between [a,b] and [c,b] is", H_line.distance((2,(1,2)),(1,(0,1)))) 
print("Distance between [b,a] and [c,b] is", H_line.distance((1,(0,1)),(1,(1,2)))) #Figure 2
```

We calculate the Wasserstein distance between $m_{|E_1,a\rangle}$ and $m_{|E_4,a\rangle}$.

```python
print(H.signed_wasserstein((0,(0,1,2)),(0,(0,1))))
```

```python
print(H.ricci((0,(0,1,2)),(0,(0,1))))
```

### Example 4
We calculate Grover Ricci curvature on line graph.

```python
G = line(5)
H_line_5 = HyperGraph(G.edges)
H_line_5.plot()
print("The Grover Ricci curvature between [a,b] and [b,a]", H_line_5.ricci((1,(1,2)),(2,(1,2))))
print("The Grover Ricci curvature between [a,b] and [b,c]", H_line_5.ricci((2,(1,2)),(3,(2,3))))
```

### Example 6
We calculate the Grover Ricci curvature on complete graph of degree $4$.

```python
from hyper_graph import comp_graph
H_comp = HyperGraph(comp_graph(4))
H_comp.plot()
print("The Grover Ricci curvature between [a,b] and [b,a]", H_comp.ricci((0,(0,1)),(1,(0,1))))
print("The Grover Ricci curvature between [a,b] and [b,c]", H_comp.ricci((1,(0,1)),(2,(1,2))))
```
